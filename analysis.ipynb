```python
%%sql 
drop table if exists results;
create table results as select * from read_json_auto('cleaned_results.json', format="array", 
    columns = {
    place: 'VARCHAR',
    class:'VARCHAR',
    name:'VARCHAR',
    gender:'VARCHAR',
    ag:'VARCHAR',
    club:'VARCHAR',
    time:'VARCHAR',
    gap:'VARCHAR',
    laps:'INT',
    awared:'VARCHAR',
    lap_time:'VARCHAR',
    lap_time_secs:'INT',
    total:'VARCHAR',
    total_secs:'INT'}
    )
```

```python
%%sql
select * from results;
```

```python
%%sql
with cte as (select distinct place, class, name, laps from results)
    select class, count(), avg(laps), max(laps),min(laps) from cte group by class
```
